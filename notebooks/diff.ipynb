```python
import numpy as np
import pandas as pd
import vectorbtpro as vbt

from vctr.data.data_loader import get_data

# Set dark theme
vbt.settings.set_theme('dark')
```

```python
data = get_data('NEAR', '30m', '2023-03-01')
```

```python
price = data['close']

ex = vbt.LEXLB.run(price, 0.02, 0.02)
entries = ex.labels == -1
exits = ex.labels == 1

pf = vbt.Portfolio.from_signals(price, entries, exits, freq='30t')
print(f'Return: {round(pf.total_return * 100)}%')
print(ex.labels.value_counts(normalize=True))

ex.plot().show()
```

```python
diffed = data.close.vbt.diff()

diffed.vbt.plot().show()
```
